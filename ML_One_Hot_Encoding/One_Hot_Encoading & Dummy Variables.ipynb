# One Hot Encoding With Scikit-Learn 

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse=False,drop = "first")
```

```python
oh_enc_arr = oh_enc.fit_transform(tips_df[['sex','smoker','day','time']])
oh_enc_arr
```
